# Speed criterion with non-smooth dynamics

Approach: before the first two words are read, the system, consisting only of links, should approach the origin, a stable fixed point. After the first two words are read, the links between them (but not between them and not-yet-read words) are "activated," meaning the standard LV dynamics are allowed to take over. Once the (effectively two-word) system gets close to a fixed point (velocity criterion), the next word is input, "activating" its links to the other words.

This setup has the benefit of having a fixed-dimensional state space and a simple criterion for moving on to the next word. An open question is whether and how anticipatory processing works. The plan now is to monitor the maximal eigenvalues of the Jacobian matrix of all of the relevant fixed points to look for evidence of shifting stability before new words are input. This is important, because the anticipation is a must-have for a parsing model.

```python
# Setting up a simple, two-word system for testing
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt


def dyn(x, growth_rates):#, sign):
    W = np.array([[1, 2], [2, 1]])
    #W = np.array([[1, 1.75], [1.25, 1]])
    #dxdt = match * x * (growth_rates - W @ x)
    dxdt = x * (growth_rates - W @ x)
    return dxdt


def speed(x, growth_rates):
    vel = dyn(x, growth_rates)#, match)
    speed = np.sqrt(vel @ vel)
    return speed


max_time = 5000
nlinks = 2
#growth0 = np.array([-0.1]*nlinks)
growth0 = np.array([-0.75, -0.65])
#growth1 = np.array([1.]*nlinks)
growth1 = np.array([0.75, 0.5])
#match = np.array([0.75, 0.1])

x = np.zeros((max_time, nlinks))
x[0,] = np.array([0.5, 0.5])
speeds = np.zeros(max_time)

speed_threshold = 0.01
tau = 0.01
curr_growth = growth0
min_time = 100
rt0, rt1 = 0, 0
sign0 = -1.
for t in range(1, max_time):
    x[t,] = x[t-1,] + tau * dyn(x[t-1,], curr_growth)
    x[t,] = np.clip(x[t,], -0.01, 1.01)
    speeds[t-1] = speed(x[t,], curr_growth)
    if curr_growth is growth0 and speeds[t-1] < speed_threshold and t > min_time:
        curr_growth = growth1
        rt0 = t
    elif curr_growth is growth1 and speeds[t-1] < speed_threshold and t > rt0+min_time:
        rt1 = t
        break


print('RT_word0: {}\nRT_word1: {}'.format(rt0, rt1))
y = x[~np.all(x == 0, axis=1)]
plt.plot(np.trim_zeros(speeds))
for i in range(nlinks):
    plt.plot(y[:,i])
plt.legend(['Speed', 'x', 'y'])
plt.ylim(0, 1.1)
plt.show()
print(y[-1,])
```
